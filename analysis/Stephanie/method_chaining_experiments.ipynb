# Working on Method Chaining

In this script I explore method chaining.

```python
import pandas as pd
import numpy as np

red_wine_df = (
        pd.read_csv("../../data/raw/winequality-red.csv", sep=";")
        .rename(columns={"fixed acidity": "fa", "volatile acidity": "va", "citric acid": "ca", "residual sugar": "rs", "free sulfur dioxide": "fsd", "total sulfur dioxide": "tsd"})
        .drop_duplicates()
        .dropna(axis=0, subset=['fa','va','ca','rs','chlorides','fsd','tsd','density','pH','sulphates','alcohol'])
        .reset_index()
    )
    
red_wine_df.head()
```
